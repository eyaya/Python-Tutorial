```python
Data=[]
for j in range(1,29):
    url = f'https://www.realethio.com/property-type/apartment-for-rent/page/{j}/'
    req = requests.get(url)
    page = BeautifulSoup(req.content,"html.parser")
    body = page.find_all("div", class_="item-body")
    
    for i in range(len(body)):
        title = body[i].find('h2', class_='item-title').find('a').text
        address=body[i].find('address', class_='item-address').text
        N_beds=int(body[i].find('ul',class_="item-amenities")\
        .find('li',class_='h-beds').find('span',class_='hz-figure').text)

        N_area=float(body[i].find('ul',class_="item-amenities")\
        .find('li',class_='h-area').find('span',class_='hz-figure').text)
        N_areaP=body[i].find('ul',class_="item-amenities")\
        .find('li',class_='h-area').find('span',class_='area_postfix').text
        Type=body[i].find('ul',class_="item-amenities")\
        .find('li',class_='h-type').find('span').text
        price = body[i].find('ul', class_="item-price-wrap").text

        Data.append([title, address, N_beds, N_area,\
                     N_areaP,Type,price])
```
